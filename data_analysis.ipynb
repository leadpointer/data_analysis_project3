```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patheffects as path_effects

# Load the dataset (make sure the path is correct)
data = pd.read_csv("/content/drive/MyDrive/DATA_PUMA/PUMA_SALES_DATASET.csv", encoding="latin")

# Convert 'Invoice Date' to datetime if it's not already
data['Invoice Date'] = pd.to_datetime(data['Invoice Date'])

# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# Create the line plot
plt.figure(figsize=(10,6))
sns.lineplot(x='Invoice Date', y='Total Sales', data=data, linestyle='-', color='b',path_effects=[path_effects.SimpleLineShadow(),path_effects.Normal()])

# Add titles and labels
plt.title('Total Sales Over Time', fontsize=16, fontweight='bold')
plt.xlabel('Invoice Date')
plt.ylabel('Total Sales')

# Create the table and assign it to the variable 'table'
table = plt.table(cellText=data.head(10).values, colLabels=data.columns, loc='bottom',bbox=[0, -1.0, 4, 0.8])

# Rotate x-axis labels for better readability
table.auto_set_font_size(False)  # Disable auto scaling of font size
table.set_fontsize(10) # Set font size to 8

# Auto-fit column widths and row heights
table.auto_set_column_width(col=[0, 1])  # Specify column indices to auto-fit
# For row height, you need to manually adjust since there isn't a direct method for row heights
for key, cell in table._cells.items():
    cell.set_height(0.05)  # Set a standard height for rows, adjust as needed
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patheffects as path_effects

# Load the dataset (make sure the path is correct)
data = pd.read_csv("/content/drive/MyDrive/DATA_PUMA/PUMA_SALES_DATASET.csv", encoding="latin")

# Convert 'Invoice Date' to datetime if it's not already
data['Invoice Date'] = pd.to_datetime(data['Invoice Date'])

# Extract year and month from 'Invoice Date'
data['Year-Month'] = data['Invoice Date'].dt.to_period('M')

# Group by 'Year-Month' and aggregate total sales
data_monthly = data.groupby('Year-Month')['Total Sales'].sum().reset_index()

# Convert 'Year-Month' back to datetime for plotting
data_monthly['Year-Month'] = data_monthly['Year-Month'].dt.to_timestamp()

# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# Create the line plot
plt.figure(figsize=(10,6))
sns.lineplot(x='Year-Month', y='Total Sales', data=data_monthly, linestyle='-', color='#7C00FE', path_effects=[path_effects.SimpleLineShadow(), path_effects.Normal()])

# Add titles and labels
plt.title('Total Sales Over Time by Month', fontsize=16, fontweight='bold')
plt.xlabel('Month')
plt.ylabel('Total Sales')

# Create the table and assign it to the variable 'table'
table = plt.table(cellText=data_monthly.head(5).values, colLabels=data_monthly.columns, loc='bottom', bbox=[0, -1.0, 1, 0.8])

# Rotate x-axis labels for better readability
table.auto_set_font_size(False)  # Disable auto scaling of font size
table.set_fontsize(10)  # Set font size to 10

# Auto-fit column widths and row heights
table.auto_set_column_width(col=[0, 1])  # Specify column indices to auto-fit
# For row height, you need to manually adjust since there isn't a direct method for row heights
for key, cell in table._cells.items():
    cell.set_height(0.05)  # Set a standard height for rows, adjust as needed
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data=pd.read_csv("/content/drive/MyDrive/DATA_PUMA/PUMA_SALES_DATASET.csv",encoding="latin")

# Select only numeric columns for correlation calculation
numeric_data = data.select_dtypes(include=['number'])

plt.figure(figsize=(10,6))
corr = numeric_data.corr()  # Calculate correlation on numeric data
sns.heatmap(corr, annot=True, cmap='Greens')
plt.title('Correlation Matrix for Puma Sales Data')
plt.show()
```
